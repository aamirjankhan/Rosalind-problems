```python
import numpy as np    
def Consensus_and_Profile(filename):
    sequences= []
    with open(filename ,'r') as fr:
        lines= fr.readlines()
        for line in lines:
            if '>' not in line:
                sequences.append(list(line.rstrip()))
    array= np.array(sequences)
    profile= {'A':0, 'C':0, 'G':0, 'T':0}
    concensus= list()
    li= list()
    li.append(['A:','C:','G:','T:'])
    num= array.shape[0]
    flag= 0
    for cell in np.nditer(array, order='F'):
        profile[str(cell)]+= 1
        flag+= 1
        if flag == num:
            max_nucleotide = max(profile, key=profile.get)
            concensus.append(max_nucleotide)
            li.append(list(profile.values()))
            profile = dict.fromkeys(profile, 0)
            flag= 0
    profile1= np.array(li)
    profile1= np.transpose(profile1)
    
    print("".join(concensus),"\n", profile1)
    
```

```python
print("Zill-E-Huma \nSP17-BSI-016")
seq='''ATCCAGCT\nGGGCAACT\nATGGATCT\nAAGCAACC\nTTGGAACT\nATGCCATT\nATGGCACT'''
print("\t\t\t\tDNA String\t\t\t\t")
print(seq,"\n")
for i in range(0,7):
    matrix=seq.split('\n')

A,T,C,G,consensus=[],[],[],[],[];
acount,tcount,gcount,ccount=0,0,0,0;
for j in range (0, len(matrix[0])):
    for i in range(0,len(matrix[0])-1):
        if (matrix[i][j]=='A'):
            acount+=1;
        elif (matrix[i][j]=='T'):
            tcount+=1;
        elif(matrix[i][j]=='C'):
            ccount+=1;
        elif(matrix[i][j]=='G'):
            gcount+=1;
    A.append(acount);
    T.append(tcount);
    C.append(ccount);
    G.append(gcount);
    if(acount>tcount and acount>gcount and acount>ccount):
        consensus.append('A');
    elif(tcount>acount and tcount>gcount and tcount>ccount):
        consensus.append('T');
    elif(gcount>acount and gcount>tcount and gcount>ccount):
        consensus.append('G')
    elif(ccount>acount and ccount>tcount and ccount>gcount):
        consensus.append('C')
    else:
        pass
    acount = 0;
    tcount = 0;
    ccount = 0;
    gcount = 0;

print("Profile: \tA",A,"\n\t\t\tT",T,"\n\t\t\tC",C,"\n\t\t\tG",G)
print("\nConsensus Sequence:",consensus)

```
